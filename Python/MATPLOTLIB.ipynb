```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,1000)
y=np.sin(x)
z=np.cos(x)
a=np.sin(x+2)
b=np.cos(x+2)
plt.plot(x,y)
```

```python
plt.scatter(x,y,color='red')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x,y,'-b')
plt.plot(x,z,'--r')
plt.plot(x,a,'-.c')
plt.plot(x,b,':k')
```
