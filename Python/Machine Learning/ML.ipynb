```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]
```

```python
plt.scatter(x,y,color='r')
plt.show()
```

```python
plt.hist(x,x,color='g')
plt.hist(y,y,color='b')
plt.show()
```

```python
plt.bar(x,y,color='k')
plt.show()
```

```python
plt.pie(x,startangle=90,labels=y)
```
